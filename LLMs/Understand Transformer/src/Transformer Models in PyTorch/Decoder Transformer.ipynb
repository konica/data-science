# From original to decoder-only transformer






## Masked self-attention/causal attention




```python
import torch

seq_len = 4
tgt_mask = (1 - torch.triu(torch.ones(1, seq_len, seq_len), diagonal=1)).bool()
print(tgt_mask)
```

To simplify implementing mask, we can use `torch.tril` instead of `torch.triu`

```python
another_mask = torch.tril(torch.ones(1, seq_len, seq_len)).bool() == 1
print(another_mask)
```
